```python
from ctransformers import AutoModelForCausalLM
```

```python
def colab_print(text, max_width=120):
    words = text.split()
    line = ""
    for word in words:
        if len(line) + len(word) + 1 > max_width:
            print(line)
            line = ""
        line += word + " "
    print(line)
```

```python
llm = AutoModelForCausalLM.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.1-GGUF", model_file="mistral-7b-instruct-v0.1.Q5_K_M.gguf", model_type="mistral")
```

```python
colab_print(llm("Who is Trump?", \
                max_new_tokens=2048, temperature=0.00001, top_k=55, top_p=0.9, repetition_penalty=1.2))
```

```python
ending_prompt = "What did I order? Answer with a list of coma separated values including \
    how many of each item I ordered."
```

```python
colab_print(llm(
    "Give me 1 big mac and 1 chilli chicken.\
    What did I order? Answer with in a jason format, which includes 3 fileds: food_name, quantity, comment", \
                max_new_tokens=2048, temperature=0.0, top_k=55, top_p=0.9, repetition_penalty=1.2))
```
